```python
import pandas as pd
df = pd.read_excel('Mental_Health_Factors.xlsx')
df['Age_Decade'] = (df['Age'] // 10) * 10

high_pressure_by_decade = df[df['Work_Pressure'] == df['Work_Pressure'].max()]['Age_Decade'].value_counts().sort_index()
high_pressure_by_decade.plot(kind="bar")
```
